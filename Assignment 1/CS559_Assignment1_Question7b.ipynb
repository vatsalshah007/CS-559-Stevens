```python
x_features = data.iloc[:, :2].values
y_output = data.iloc[:, -1].values
```

### Converting the output values from strings to values.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_output = label_encoder.fit_transform(y_output)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X = train_test_split(x_features, test_size = 0.30)
train_Y, test_Y = train_test_split(y_output, test_size = 0.30)
```

## Finding the Gradient Descent

```python
weights = np.zeros((2, 1))
```

```python
learningRate = 0.001
iterator = 1000
N = len(x_features)
weights = np.random.rand(2, 1)
bias = 0

for i in range(iterator):
  prediction = 1/(1+1/np.exp(np.dot(weights.T, train_X.T) + bias))

cost = -(1/N)*np.sum(train_Y*np.log(prediction) + (1 - train_Y)*np.log(1-prediction))

differenceWeights = 1/N*np.dot(train_X.T, (prediction-train_Y).T)
differenceBias = 1/N*np.sum(prediction-train_Y)

weights = weights - learningRate*differenceWeights
bias = bias - learningRate*differenceBias
```
